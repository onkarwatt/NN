- Write a Python program to plot a few activation functions that are being used in neural networks. 

```python
import matplotlib.pyplot as plt
import numpy as np


def plot_sigmoid():
    x = np.linspace(-10,10,100)
    y = 1 / (1+np.exp(-x))
    
    plt.plot(x,y)
    plt.xlabel('Input')
    plt.ylabel('sigmoid output')
    plt.title('sigmoid activation function')
    plt.grid(True)
    plt.show()

plot_sigmoid()

def tanh():
    x = np.linspace(-10,10,100)
    tanh = np.tanh(x)
    
    plt.plot(x,tanh)
    plt.title("hyperbolic tangent activation function")
    plt.xlabel('X')
    plt.ylabel('tanh(x)')
    plt.grid(True)
    plt.show()

tanh()

def plot_relu():
    x = np.linspace(-10,10,100)
    relu = np.maximum(0,x)
    
    plt.plot(x,relu)
    plt.title("Relu activation function")
    plt.xlabel("x")
    plt.ylabel("relu(x)")
    plt.grid(True)
    plt.show()

plot_relu()

def plot_leaky_relu():
    x = np.linspace(-10,10,100)
    
    def leaky_relu(x,alpha = 0.1):
        return np.where(x >= 0,x,alpha*x)
    
    leaky_relu_values = leaky_relu(x)
    
    plt.plot(x,leaky_relu_values)
    plt.title("leaky relu activation function")
    plt.xlabel('x')
    plt.ylabel("leaky_relu(x)")
    plt.grid(True)
    plt.show()

plot_leaky_relu()

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / np.sum(e_x, axis=0)

def plot_softmax(probabilities, class_labels):
    plt.bar(class_labels, probabilities)
    plt.xlabel("Class")
    plt.ylabel("Probability")
    plt.title("Softmax Function")
    plt.show()

class_labels = ["Class A", "Class B", "Class C"]
probabilities = softmax(np.array([0.2, 0.3, 0.5]))

print(probabilities)
plot_softmax(probabilities,class_labels)
```
